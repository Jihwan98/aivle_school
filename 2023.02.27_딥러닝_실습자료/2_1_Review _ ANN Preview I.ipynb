```
from sklearn.datasets import fetch_california_housing
```

```
california = fetch_california_housing()
```

```
x = california.data
y = california.target
```

```
x.shape, y.shape
```

```
import pandas as pd

pd.DataFrame(x, columns=california.feature_names).head(2)
```

```
pd.DataFrame(y, columns=california.target_names).head(2)
```

```
#####################
# 라이브러리 불러오기
#####################
import tensorflow as tf
from tensorflow import keras
```

```
########
# 모델링
########
# 1. 세션 클리어
keras.backend.clear_session()

# 2. 모델 생성
model = keras.models.Sequential()

# 3. 모델 쌓기
model.add( keras.layers.Input(shape=(8,)) )
model.add( keras.layers.Dense(1) )

# 4. 모델 컴파일
model.compile(loss='mse', optimizer='adam')
```

```
###########
# 모델 학습
###########
model.fit(x, y, epochs=10)
```

```
###########
# 모델 예측
###########
y_pred = model.predict(x)
print(y_pred.reshape(-1)[:5])
print(y[:5])
```

# 히든 레이어 추가하여 모델링!
---
- 조건
    - 히든 레이어 2개를 추가하여 다시 모델링!
    - 히든 레이어 노드 수 : 32
    - 히든 레이어 activation = 'relu'

```
########
# 모델링
########
# 1. 세션 초기화
keras.backend.clear_session()

# 2. 모델 생성
model = keras.models.Sequential()

# 3. 모델 쌓기
model.add( keras.layers.Input(shape=(8, )) )
model.add( keras.layers.Dense(32, activation='relu') )
model.add( keras.layers.Dense(32, activation='relu') )
model.add( keras.layers.Dense(1))

# 4. 모델 컴파일
model.compile(loss='mse', optimizer='adam')

# 모델 summary
model.summary()
```

```
###########
# 모델 학습
###########
model.fit(x, y, epochs=10)
```

```
###########
# 모델 예측
###########
y_pred = model.predict(x)
print(y_pred.reshape(-1)[:5])
print(y[:5])
```
