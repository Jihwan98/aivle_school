```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
x = iris.data
y = iris.target
```

```python
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=2023)
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
#####################
# 라이브러리 불러오기
#####################
import tensorflow as tf
from tensorflow import keras
from keras.utils import to_categorical
import numpy as np
```

```python
class_n = len(np.unique(y_train))
class_n
```

```python
#######################
# 어떤 전처리 작업 필요
# Hint : y.shape
#######################
y_train = to_categorical(y_train, class_n)
y_train.shape
```

```python
y_train[:5]
```

```python
########
# 모델링
########
# 1. 세션 초기화
keras.backend.clear_session()

# 2. 모델 생성
model = keras.models.Sequential()

# 3. 모델 쌓기
model.add( keras.layers.Input(shape=(4, )) )
model.add( keras.layers.Dense(3, activation='softmax') )

# 4. 모델 컴파일
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')

# 모델 summary
model.summary()
```

```python
###########
# 모델 학습
###########
model.fit(x_train, y_train, epochs=10)
```

```python
###########
# 모델 예측
###########
y_pred = model.predict(x_test)
print(y_pred[:5])
print(y_test[:5])
```

```python
print(np.array(list(map(lambda x: np.argmax(x), y_pred))))
print(y_test)
```

# 히든 레이어 추가하여 모델링! (Functional API)
---
- 조건
    - 히든 레이어 2개를 추가하여 다시 모델링!
    - 히든 레이어 노드 수 : 32
    - 히든 레이어 activation = 'relu'

```python
########
# 모델링
########
# 1. 세션 초기화
keras.backend.clear_session()

# 2. 모델 생성
model = keras.models.Sequential()

# 3. 모델 쌓기
model.add( keras.layers.Input(shape=(4, )) )
model.add( keras.layers.Dense(32, activation='relu') )
model.add( keras.layers.Dense(32, activation='relu') )
model.add( keras.layers.Dense(3, activation='softmax') )

# 4. 모델 컴파일
model.compile(loss='categorical_crossentropy', metrics='accuracy', optimizer='adam')

# 모델 summary
model.summary()
```

```python
###########
# 모델 학습
###########
model.fit(x_train, y_train, epochs=10)
```

```python
###########
# 모델 예측
###########
y_pred = model.predict(x_test)
print(y_pred[:5])
print(y_test[:5])
```

```python
print(np.array(list(map(lambda x: np.argmax(x), y_pred))))
print(y_test)
```
