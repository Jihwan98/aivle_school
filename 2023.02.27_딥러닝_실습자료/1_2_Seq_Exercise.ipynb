```
import tensorflow as tf
from tensorflow import keras

import numpy as np
import pandas as pd
```

##Linear Regression on 쬐금 더 어려운 데이터

```
from sklearn.datasets import load_boston
```

```
boston = load_boston()
```

```
x = boston.data
y = boston.target

x.shape, y.shape
```

```
## Sequential
# 1. 세션 초기화
keras.backend.clear_session()

# 2. 모델 발판
model = keras.models.Sequential()

# 3. 모델 쌓기
model.add(keras.layers.Input(shape=(13,)))
model.add(keras.layers.Dense(1))

# 4. 컴파일
model.compile(loss='mse', optimizer='adam')
```

```
# 5. 학습
model.fit(x, y, epochs=10)
```

```
# 6. 예측
y_pred = model.predict(x)
print("Predict :", y_pred.reshape(-1)[:5])
print("Actual :", y[:5])
```

```
boston = pd.read_csv('https://raw.githubusercontent.com/DA4BAM/dataset/master/boston.csv')
target = 'medv'
x = boston.drop(target, axis=1)
y = boston.loc[:, target]

x.shape, y.shape
```

```
## Sequential API
# 1. 세션 지우기
keras.backend.clear_session()

# 2. 모델 발판
model = keras.models.Sequential()

# 3. 모델 쌓기
model.add( keras.layers.Input(shape=(12, )) )
model.add( keras.layers.Dense(1) )

# 4. 컴파일
model.compile(loss=keras.losses.mse, optimizer=keras.optimizers.Adam())
```

```
# 5. 모델 학습
model.fit(x, y, epochs=10)
```

```
# 6. 예측
y_pred = model.predict(x)
print("Predict :", y_pred.reshape(-1)[:5])
print("Actual :", y.values[:5])
```

## Logistic Regression 쬐금 더 어려운 데이터

```
from sklearn.datasets import load_breast_cancer
```

```
cancer = load_breast_cancer()
```

```
x = cancer.data
y = cancer.target

# 모양 확인해볼 것
x.shape, y.shape
```

```
print(cancer.DESCR)
```

```
## Sequential API
# 1. 세션 초기화
keras.backend.clear_session()

# 2. 모델 발판 생성
model = keras.models.Sequential()

# 3. 모델 쌓기
model.add( keras.layers.Input(shape=(30,)) )
model.add( keras.layers.Dense(1, activation='sigmoid') )

# 4. 모델 컴파일
model.compile(loss='binary_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```
# 5. 모델 학습
model.fit(x, y, epochs=10)
```

```
# 6. 예측
y_pred = model.predict(x)

print(y_pred.reshape(-1)[:5])
print(y[:5])
```
