```python
import tensorflow as tf
from tensorflow import keras
```

```python
# 1. 세션 클리어
keras.backend.clear_session()

# 2. 레이어 연결
il = keras.layers.Input(shape=(5, 5, 3))
cl = keras.layers.Conv2D(filters=16, kernel_size=(3, 3), strides=1, activation='relu')(il)
pl = keras.layers.MaxPool2D(strides=1)(cl)

# 3. 모델의 시작과 끝 지정
model = keras.models.Model(il, pl)

model.summary()
```

```python
x = tf.constant([[1., 2., 3.],
                 [4., 5., 6.],
                 [7., 8., 9.]])
x = tf.reshape(x, [1, 3, 3, 1])
max_pool_2d = tf.keras.layers.MaxPooling2D(pool_size=(2, 2),
   strides=(2, 2), padding='same')
max_pool_2d(x)
```
