```python
import gradio as gr
ko2enDict = {
    "사과" : "apple",
    "사자" : "lion",
    "사랑" : "love"
}
def trans_ko_en(ko):
    return ko2enDict[ko]
app = gr.Interface(
    fn=trans_ko_en,
    inputs=gr.Textbox(placeholder="한국어"),
    outputs="text",
)
app.launch(debug=True)
```

```python
import gradio as gr
dict = {
    "ko2en" : {
        "사과" : "apple",
        "사자" : "lion",
        "사랑" : "love"
    },
    "en2ko" : {
        "apple" : "사과",
        "lion" : "사자",
        "love" : "사랑"
    }
}
def trans(type, word):
    return dict[type][word]
app = gr.Interface(
    fn=trans,
    inputs=[
        gr.Radio(['ko2en', 'en2ko']),
        gr.Textbox(placeholder="한국어")
    ],
    outputs="text",
)
app.launch(debug=True)
```

```python
import gradio as gr
dict = {
    "ko2en" : {
        "사과" : "apple",
        "사자" : "lion",
        "사랑" : "love"
    },
    "en2ko" : {
        "apple" : "사과",
        "lion" : "사자",
        "love" : "사랑"
    }
}
def trans(type, word):
    return dict[type][word]
with gr.Blocks() as app:
    result = gr.Textbox(label='결과')
    type = gr.Radio(label='종류', choices=['ko2en', 'en2ko'])
    word = gr.Textbox(label='단어', placeholder="한국어")
    btn = gr.Button(value='실행')
    btn.click(
        fn = trans,
        inputs = [type, word],
        outputs = result,
        api_name = 'trans'
    )
app.launch()
```

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis") # 모델 다운로드
classifier(["I've been waiting for a HuggingFace course my whole life.",
            "Good!",
            "bad"])
```

```python
import gradio as gr
from transformers import pipeline

def sentiment(sentence):
    classifier = pipeline("sentiment-analysis")
    return classifier(sentence)

with gr.Blocks() as app:
    result = gr.Textbox(label='결과')
    sentence = gr.Textbox(label='분석할 문장')
    btn = gr.Button(value='실행')
    btn.click(
        fn = sentiment,
        inputs = sentence,
        outputs = result
    )
app.launch()
```
