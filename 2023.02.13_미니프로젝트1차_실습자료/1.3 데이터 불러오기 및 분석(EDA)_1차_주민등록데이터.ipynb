#### [실습문제4] 데이터 분포 알아보기
* 다양한 변수를 기준으로 그래프를 그려보고 인사이트를 도출해보세요.

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
plt.figure(figsize=(12, 6))
for i in range(1, len(list(seoul_people))):
    plt.subplot(1, 2, i)
    plt.title(list(seoul_people)[i])
    plt.bar(seoul_people['자치구'], seoul_people[list(seoul_people)[i]])
    plt.xticks(rotation=90)
    
plt.tight_layout()
plt.show()
```

-  위 차트를 통해 알게된 사실을 정리해봅시다.
    1. 세대, 인구, 한국인은 송파구에 제일 많다.
    2. 금천구는 인구가 적다.
    3. 구로구, 영등포구에는 외국인이 많다.
